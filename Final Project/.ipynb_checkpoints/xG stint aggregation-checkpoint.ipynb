## assign stint ID to each shoot, and record the max time within stint

```python
l = []

for index, row in shoot.iterrows():
    Match_Half = row['Match_Half']
    Minute = row['Minute']
    MatchID = row['matchID']
    
    target = stint[(stint['matchID'] == MatchID) & (stint['Event_Half'] == Match_Half)]
    flag = False
    
    for i in range(len(target) - 1):
        if (Minute > target.iloc[i]['time_start']) & (Minute <= target.iloc[i+1]['time_start']):
            l.append(target.iloc[i]['EventID'])
            flag = True
            break
    
    if(flag == False):
        if Minute == 1:
            l.append(0)
            flag = True
        else:
            l.append(target.iloc[-1]['EventID'])
            flag = True
        
l
```

```python
len(l)
```

```python
shoot_stint = pd.concat([shoot, pd.DataFrame(l, columns = {'EventID'})], axis = 1)
shoot_stint
```

```python
shoot_stint = shoot_stint.merge(stint)
shoot_stint
```

## update stint length

```python
long_stint = shoot_stint[shoot_stint['Minute'] > (shoot_stint['time_start'] + shoot_stint['stint_length'])]
long_stint = long_stint.groupby(['matchID', 'EventID'])['Minute'].max()
long_stint = pd.DataFrame(long_stint).reset_index()
long_stint
```

```python
for index, row in long_stint.iterrows():
    matchID = row['matchID']
    EventID = row['EventID']
    Minute = row['Minute']
    index = stint[(stint['matchID'] == matchID) & (stint['EventID'] == EventID)].index.values[0]
    stint.iloc[index]['stint_length'] = (Minute - stint.iloc[index]['time_start'])
```

```python
shoot_stint = shoot_stint.drop(columns = ['stint_length'])
shoot_stint = shoot_stint.merge(stint)
shoot_stint
```

```python
shoot_stint[shoot_stint['Minute'] > (shoot_stint['time_start'] + shoot_stint['stint_length'])]
```

## aggregate xG for home/away team per stint

```python
stint_xg = shoot_stint.groupby(['matchID', 'EventID', 'Home_Away'])['xG'].sum()
stint_xg = pd.DataFrame(stint_xg).reset_index()
stint_xg
```

```python
stint_ha_xg = stint_xg[['matchID', 'EventID']].drop_duplicates().reset_index(drop = True)
stint_ha_xg
```

```python
h = []
a = []

for index, row in stint_ha_xg.iterrows():
    matchID = row['matchID']
    EventID = row['EventID']
    
    h.append(stint_xg[(stint_xg['matchID'] == matchID) & (stint_xg['EventID'] == EventID) & (stint_xg['Home_Away'] == 'Home')]['xG'].values)
    a.append(stint_xg[(stint_xg['matchID'] == matchID) & (stint_xg['EventID'] == EventID) & (stint_xg['Home_Away'] == 'Away')]['xG'].values)
```

```python
stint_ha_xg = pd.concat([stint_ha_xg, pd.DataFrame(h, columns = {'Home_stint_xG'})], axis = 1)
stint_ha_xg = pd.concat([stint_ha_xg, pd.DataFrame(a, columns = {'Away_stint_xG'})], axis = 1)
stint_ha_xg
```
