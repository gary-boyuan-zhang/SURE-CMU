### Check Match Order

```python
fb_order = fb[['Date', 'Squad', 'Home_Away']].drop_duplicates().reset_index().drop(columns = ['index'])
fb_order
```

```python
ha = []
for index, row in fb_order.iterrows():
    if index % 2 == 0:
        ha.append('Home')
    else:
        ha.append('Away')
        
ha
```

```python
fb_order = fb_order.drop(columns = ['Home_Away'])
fb_order = pd.concat([fb_order, pd.DataFrame(ha, columns = {'Home_Away'})], axis = 1)
fb_order
```

```python
fb_order = pd.concat([fb_order, pd.DataFrame(np.repeat(list(range(0, 380)), 2), columns = {'order'})], axis = 1)
fb_order
```

```python
us_order = us[['date', 'home_team', 'away_team', 'match_id']].drop_duplicates().reset_index().drop(columns = ['index'])
us_order
```

```python
from datetime import datetime, date

us_order['date'] = pd.to_datetime(us_order['date'])
us_order['date'] = us_order['date'].dt.date.astype(str)
#us_order['date'] = us_order['date']
us_order
```

```python
l = []
for index, row in us_order.iterrows():
    home = row['home_team']
    date = str(row['date'])
    l.append(fb_order[(fb_order['Squad'] == home) & (fb_order['Home_Away'] == 'Home') 
                      & (fb_order['Date'] == date)]['order'].values.astype(int))
    
l
```

```python
us_order = pd.concat([us_order, pd.DataFrame(l, columns = {'order'})], axis = 1)
us_order
```

### find diff. 9 obs.

```python
agg_fb = fb.groupby(['order'])['Date'].count()
agg_fb
```

```python
agg_us = us.groupby(['order'])['id'].count()
agg_us
```

```python
count = 0
c = 0
uf = []
fu = []
for i in range(380):
    if agg_fb[i] != agg_us[i]:
        if agg_us[i] > agg_fb[i]:
            uf.append((i, agg_fb[i], agg_us[i]))
        else:
            fu.append((i, agg_fb[i], agg_us[i]))
        
```

### remove ownGoal

```python
us = us[us['result'] != 'OwnGoal']
us
```

### fix time diff. in US

```python
us['minute'] = us['minute'] + 1
us
```

### fix uf

```python
del_id = []
for index, row in uf.iterrows():
    order = row[0]
    fb_match = fb[fb['order'] == order]
    us_match = us[us['order'] == order]
    
    flag = False
    for i in range(len(fb_match)):
        if us_match.iloc[i]['minute'] != fb_match.iloc[i]['Minute']:
            del_id.append(us_match.iloc[i]['id'])
            flag = True
            break
    
    if(flag == False):
        del_id.append(us_match.iloc[len(fb_match)]['id'])
        flag = True
      
    
    
del_id
```

```python
len(del_id)
```

```python
us = us[~(us['id'].isin(del_id))]
us
```

### fix fu

```python
fu = pd.DataFrame(fu)
fu[3] = fu[1] - fu[2]
fu
```

```python
fu[3].sum()
```

```python
del_index = []
for index, row in fu.iterrows():
    order = row[0]
    fb_match = fb[fb['order'] == order]
    us_match = us[us['order'] == order]
    diff = row[1] - row[2]
    index_match = fb_match.index.tolist()
    
    count = 0
    for i in range(len(us_match)):
        if us_match.iloc[i - count]['minute'] != fb_match.iloc[i]['Minute']:
            del_index.append(index_match[i])
            count += 1
            if count == diff:
                break
         
    
    if(count < diff):
        for j in range(1, diff - count + 1):
            del_index.append(index_match[-j])
            count += 1
      
    
    
del_index
```

```python
len(del_index)
```

```python
fb = fb.drop(del_index)
fb
```
