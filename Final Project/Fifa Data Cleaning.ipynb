```python
import pandas as pd
import numpy as np
```

```python
pl_cb = pd.read_csv("pl_cb.csv")
pl_cb
```

```python
fifa22 = pd.read_csv("FIFA22_official_data.csv")
fifa22
```

```python
pl_clubs = pl_cb['Squad'].unique()
fifa22_pl = fifa22[fifa22['Club'].isin(pl_clubs)]
fifa22_pl
```

```python
d = {}

for index, player in pl_cb.iterrows():
    ln = player['Player'].split()[-1]
    #fn = player['Player'].split()[0]
    temp = fifa22[(fifa22['Name'].str.contains(ln))] #| (fifa22['Name'].str.contains(fn))]
    
    if len(temp) == 1:
        d[player['Player']] = temp['ID'].values
    elif len(temp) == 0:
        d[player['Player']] = 'NA'
    else:
        temp = temp[(temp['Age'] == player['Age']) | (temp['Club'] == player['Squad'])]
        
        if len(temp) == 1:
            d[player['Player']] = temp['ID'].values
        else:
            d[player['Player']] = 'MP'
```

```python
d
```

```python
d['Pontus Jansson'] = np.array([194201])
d['Lewis Dunk'] = np.array([199915])
d['Nathan Collins'] = np.array([248484])
d['Michael Keane'] = np.array([207599])
d['Yerry Mina'] = np.array([220523])
d['Jonny Evans'] = np.array([169588])
d['Wesley Fofana'] = np.array([248695])
d['Rúben Dias'] = np.array([239818])
d['Aymeric Laporte'] = np.array([212218])
d['Davinson Sánchez'] = np.array([220793])


d['Gabriel Dos Santos'] = np.array([232580])
```

```python
player_name = pd.DataFrame.from_dict(d, orient = 'index').reset_index().rename(columns = {'index': 'Player', 0: 'ID'})
player_name
```

```python
pl_cb_fifa22 = pd.merge(pl_cb, player_name, on = 'Player')
pl_cb_fifa22 = pd.merge(pl_cb_fifa22, fifa22[['ID', 'Overall']], on = 'ID')
pl_cb_fifa22
```

```python
pl_cb_fifa22.to_csv('pl_cb_fifa22.csv', index = False)
```
